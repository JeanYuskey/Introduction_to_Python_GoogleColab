#Variáveis
Variáveis são um dos recursos das linguagens de programação que são utilizadas para armazenar valores em memória, elas nos permitem gravar e ler esses dados com facilidade a partir de um nome definido por nós.
Os nomes de variáveis devem respeitar algumas regras:
* Case sensitive (total ≠ Total)
* Podem ter tamanho ilimitado (mas evite abusos) Devem começar com letra ou underline ( _ )
* Outros caracteres podem ser letras, números ou underline
* Não podem ter espaço nem acentos
* Não podem ser uma palavra reservada da linguagem

#Tipos de dados

Em Python, toda variável tem um tipo

> * Com isso, o computador pode saber quais operações são permitidas

> > * Os tipos podem ser divididos em três grupos

> > > *  Tipos numéricos (inteiro, float, ...)

> > > *  Tipos textuais (caractere e string)

> > > *  Tipo lógico (booleano)

> * Os tipos são definidos dinamicamente, pelo próprio Python

#OPERADORES ARITMÉTICOS:

Os **operadores aritméticos** podem ser definidos nos seguintes itens:

 OPERADOR  | SÍMBOLO   |  EXECUÇÃO
 -------------------|------------------|------------------
 ADIÇÃO	 | + | 	a + b → soma dois ou mais valores
 SUBTRAÇÃO	 | - | 	a – b → subtrai b de a
 MULTIPLICAÇÃO	 | * | 	a * b  produto de a por b
 DIVISÃO	 | / | 	a / b → divisão de a por b
 POTENCIAÇÃO	 | ** | 	a ** b → a elevado a b  potenciação de a por b
 RESTO DA DIVISÃO(MOD)	 | % | 	a % b → resto da divisão inteira de a por b
 QUOCIENTE(DIV)	 | // | 	a // b → quociente da divisão inteira de a por b




#OPERADORES ARITMÉTICOS (HIERARQUIA DAS OPERAÇÕES)
* Parênteses tem a mais alta precedência e podem ser usados para forçar uma expressão a ser avaliada na ordem de preferência;
* Exponenciação ou potenciação
* Multiplicação e divisão
* Adição e subtração
_**Operadores com a mesma precedência são avaliados da esquerda para a 	direita.**_


#OPERADORES RELACIONAIS:

Os **operadores relacionais**:

 OPERADOR | SÍMBOLO | 	EXECUÇÃO 
 -------------------|------------------|------------------
 IGUALDADE	| == | 	a == b → verifica a igualdade entre dois números
 MAIOR	| > | 	a > b → verifica se a é maior que b
 MENOR	| < | 	a < b → verifica se a é menor que b
 MAIOR OU IGUAL	| >= | 	a > = b → verifica se a é maior ou igual a b
 MENOR OU IGUAL	| <= | 	a <= b → verifica se a é menor ou igual a b
 DIFERENTE	| != | 	a != b → verifica se a e b são diferentes





##Operadores relacionais exemplo
 Operador | 	Exemplo
 -------------------|------------------
 x < y | 	5 < 3 → False
 x <= y | 	5 <= 3 → False
 x > y | 	5 > 3 → True
 x >= y | 	5 >= 3 → True
 x == y | 	5 == 3 → False
 x != y | 	5 != 3 → True

#OPERADORES LÓGICOS:
* Tratam-se de operadores que executam operações lógicas envolvendo duas ou mais expressões relacionais.

Os **operadores lógicos** são:

 símbolo | operação
--------|---------
and | (E)
$or$ | (OU)
! |(NÂO)

#Operações lógicas

EXPRESSÃO A	 | and | 	EXPRESSÃO B | 	RESULTADO
 -------------------|------------------|------------------|------------------
 False	 | and | 	False | 	False
 False	 | and | 	True | 	False
 True	 | and | 	False | 	False
 True	 | and | 	True | 	True

 EXPRESSÃO A | or | 	EXPRESSÃO B | 	RESULTADO
 -------------------|------------------|------------------|------------------
 False	| or |	False | 	False
 False	| or |	True | 	True
 True	| or |	False | 	True
 True	| or |	True | 	True

 EXPRESSÃO A	 | not | 	RESULTADO
 -------------------|------------------|------------------
 False	| not |	True
 True	| not |	False


##Operadores relacionais exemplo 
 Operador | 	Exemplo
 -------------------|------------------
 x < y | 	5 < 3 → False
 x <= y | 	5 <= 3 → False
 x > y | 	5 > 3 → True
 x >= y | 	5 >= 3 → True
 x == y | 	5 == 3 → False
 x != y | 	5 != 3 → True
 
* Prioridade sempre inferior aos operadores aritméticos
* Sempre têm **resultado booleano**


 Operador | Exemplo
 -------------------|------------------
 not x | 	not True → False
 x and y | 	True and False → False
 x or y | 	True or False → True
