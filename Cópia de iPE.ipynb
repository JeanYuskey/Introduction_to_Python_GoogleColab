###Estrutura Condicional

* Trata-se de uma estrutura responsável pelo desvio na seqüência de execução das instruções em um programa
* Executam o desvio da execução do programa em conformidade com o resultado lógico de uma expressão relacional ou lógica (CONDIÇÃO).

###Decisão

Mecanismos de decisão:

`if ...`
> * Executa algo somente quando uma condição é verdadeira
`if... else`
> * Bifurca a execução do código em função de uma condição
`if... elif...`
> * Executa apenas o bloco em que a condição é verdadeira

###Decisão do tipo if...
* Executa o bloco de instruções somente se a condição for  verdadeira
* A condição é uma expressão booleana que pode fazer  uso de quaisquer operadores
* O bloco de instruções é delimitado por indentação

Exemplo301: Programa para informar quando um número inteiro é par:

Image 095 2022 12.png

````python
numero = int(input("Entre com um número: "))  
resto=numero % 2
if resto == 0:
	print("O número é par.")
````

###Decisão do tipo if... else
* Executa um ou o outro bloco de instruções em função da  condição ser verdadeira ou falsa
* Valem as mesmas regras para `if...`
* Qualquer combinação de instrução individual ou em  bloco é aceita no corpo do `if` ou do `else`
* Podem ser aninhados com outras estruturas

Exemplo302: Programa para informar se um número é par ou impar:


Image 096 2022 12.png

````python
numero = int(input("Entre com um número: "))  
# lembrando sempre que " = " é atribuição, o mesmo que " ← " e " == " é comparação, um operador logico que compara valores e da um valor de Verdadeiro ou Falso
# Nesse caso "numero mod 2 == 0" o computador esta se perguntando se "numero % 2" é igual a 0, esta comparando um com o outro, e tera como resposta ou True ou False
if numero % 2 == 0:
	print("O número é par.")
else:
	print("O número é impar.")
````

###Decisão do tipo if... elif...

* Apenas o bloco no qual a condição é  verdadeira é executado
* É possível colocar tantos _**elif**_ quantos  forem necessários
* Qualquer combinação de instrução  individual ou em bloco é aceita no corpo  do **_`if`_** ou do _**`elif`**_
* É possível adicionar um _**`else`**_ ao final de  tudo
> Nesse caso, se nenhuma condição for  verdadeira, o bloco do else será executado

Oque esta acontecendo? Porque o codigo pode exibir 3 respostas ao mesmo tempo
> *Quando utilizamos if devemos pensar que criamos uma linha nova no nosso fluxograma onde se algo é verdade apartir de um ponto algo acontece, ao utilizarmos 3 ifs criamos 3 linhas novas e se as 3 forem verdadeiras teremos tres respostas*

Então oque fazer para que isso não ocorra nesse caso onde queremos só uma resposta?
> Devemos utilizar "if", "elif" e "else" para criar um codigo fechado, o "if" abre uma possibilidade que vai ser sempre executada se for verdadeira, utilizar "elif" faz com apenas umas dessas possibilidades seja exibida e o else sera executado quando nem o "if" ou qualquer um dos "elif" for verdadeiro



````python
n1=int(input("insira o valor de n1"))
n2=int(input("insira o valor de n2"))
n3=int(input("insira o valor de n3"))
#Então nesse caso eu utilizo "if" para fazer a primeira condição
if n1 >= n2 and n1 >= n3:
  print("Maior",n1)
#Depois utilizo o "elif" para que dentro das condições que eu criei o codigo exiba apenas uma saida 
elif n2 >= n1 and n2 >= n3:
  print("Maior",n2)
elif n3 >= n1 and n3 >= n2:
  print("Maior",n3)
````

O "if" cria uma condição e usar "elif" bifurca essa condição de forma com que eu possa utilizar quantos "elif" eu quiser mas apenas apenas a condição verdadeira sera executada, e caso haja mais de uma condição verdadeira, pois existem resultados iguais, apenas um resultado sera exibido

Indo mais um mais um pouco adiante no objetivo de sempre tentar diminuir o codigo o maximo possivel, podemos usar o "else" 

O else vai ser executado quando nem o "if" ou nenhum "elif" for verdade


---
Como usar nesse caso?
> Se o primeiro "if" é falso e o primeiro "elif" tambem, como não temos mais nenhuma condição, alem dessas 3, ao invez de escrevermos o terceiro "elif" podemos apenas utilizar o "else" pois se temos 3 opções, se não é a primeira nem a segunda, logo é a terceira


````python
n1=int(input("insira o valor de n1"))
n2=int(input("insira o valor de n2"))
n3=int(input("insira o valor de n3"))
if n1 >= n2 and n1 >= n3:
  print("Maior",n1)
elif n2 >= n1 and n2 >= n3:
  print("Maior",n2)
# Nesse caso, ao invez de escrever mais uma condição de "n3 >= n1 and n3 >= n2" apenas usei o "else" para simplificar o algoritimo
else:
  print("Maior",n3)
````

Uma outra forma de resolver é pensar que ao invez de exibir uma das variaveis que criamos podemos utilizar uma putra variavel e atribuir o valor da maior variavel a ela, fazendo um swap, esse metodo é mais simples, principalmente quando o numero de variaveis que utilizamos começa a fivar maior, evita que tenhamos que escrever linhas e linhas de codigo nesses casos

````python
n1=int(input("insira o valor de n1"))
n2=int(input("insira o valor de n2"))
n3=int(input("insira o valor de n3"))
#Determinamos nossa variavel maior e damos a ela o valor de n1 (poderia ser qualquer um dos n, mas se for n1 devemos checar os outros dois, se for n2 ou n3 a mesma coisa)
maior = n1
#Então checamos se n2 é maior que a variavel "maior", caso seja atribuimos a variavel "maior" esse valor de n2 que é maior
if maior < n2:
  maior = n2
#Checamos então se n3 é maior que a variavel "maior", e da mesma forma, caso n3 sea maior, atribuimos a variavel "maior" o valor de n3
if maior < n3:
  maior = n3
#Dessa forma a variavel "maior" tera o maior valor que foi inserido no codigo
print("O maior valor é",maior)
````

algoritimo para calcular fatorial 
*Em algoritimo escrito*

```
# algoritimo "fatorial"
var
  n,f,fat:inteiro
  // "f" é o numero que queremos realizar o fatorial o f!
inicio
  escreva("Qual o valor do f")
  leia(f)
  fat ← 1 
  n ← 1 
  enquanto n < f faça
    fat ← fat*n
    n ← n+1
  fim enquanto 
  escreva(f,"!=",fat)
fim algoritimo
  
```

Fazer um programa em Python para um caixa eletrônico. O programa deve receber como entrada um valor inteiro em reais e apresentar quantas cédulas serão fornecidas para formar este montante (cédulas de 100, 50, 20, 10, 5 e 2 reais). Por exemplo, se o usuário digitar 77 reais, a saída será:


* 1 cédula de 50

* 1 cédula de 20

* 1 cédula de 5

* 1 cédula de 2 

  

Uma abordagem é começar retirando as cédulas maiores, por meio da divisão, e depois guardando quanto sobrou. Por exemplo, 234 reais:

- 234 é maior que 100
- 234 dividido por 100 = 2 cédulas de 100
- Sobra o resto da divisão por 100 (234 % 100 = 34) para ser testado e dividido entre as demais cédulas.


````python
print(11 % 2)
````

````python
c100 = 0
c50 = 0
c20 = 0
c10 = 0
c5 = 0
c2 = 0
valor = int(input("Digite o valor que deseja depositar"))
if valor >= 100:
  c100 = valor // 100
  print(c100," cedulas de 100")
  valor %= 100
if valor >= 50:
  c50 = valor // 50
  print(c50," cedulas de 50")
  valor %= 50
if valor >= 20:
  c20 = valor // 20
  print(c20," cedulas de 20")
  valor %= 20
if valor >= 10:
  c10 = valor // 10
  print(c10," cedulas de 10")
  valor %= 10
if valor >= 5:
  c5 = valor // 5
  print(c5," cedulas de 5")
  valor %= 5
if valor >= 2:
  c2 = valor // 2
  print(c2," cedulas de 2")
  valor %= 2
if valor != 0:
  print("sobrou",valor,"R$, esse valor sera devolvido em moeda")
else:
  print("não existe troco")
#esse programa arruma de maneira podre o problema, um modo mehor seria fazer com que a distribuição de notas 5 e 2
````
