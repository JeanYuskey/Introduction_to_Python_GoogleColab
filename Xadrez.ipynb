````python

tab=[[' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' '],
     [' ',' ',' ',' ',' ',' ',' ',' ']]

tab[5][3]='b'
tab[2][4]='t'
tab[7][5]='t'
tab[6][2]='q'
def mostratab(atuali,atualj,simbolo):
  print("posição i:{0} j:{1}".format(atuali,atualj))
  for i in range(8):
    for j in range(8):
      if atuali==i and atualj==j:
        print(simbolo,end="")
      else:
        if tab[i][j]==' ':
          print('.',end="")
        else:
          print(tab[i][j],end="")
    print()
def vBispo(pi,pj,peca):
  direcaoi=(-1,-1, 1,1)
  direcaoj=(-1, 1,-1,1)
  limitei=(-1, -1, 8,8)
  limitej=(-1, 8, -1,8)
  ok=False
  for k in range(4):
    i=pi
    j=pj
    while True:
      if tab[i][j]==peca:
        ok=True
        mostratab(i,j,"x")
        break
      i+=direcaoi[k]
      j+=direcaoj[k]
      if i==limitei[k] or j==limitej[k]:
        break
    if ok:
      break
  return ok

def vTorre(pi,pj,peca):
  direcaoi=(-1, 0, 0,1)
  direcaoj=( 0,-1, 1,0)
  limitei=(-1,-1,-1,8)
  limitej=(-1,-1, 8,-1)
  ok=False
  for k in range(4):
    i=pi
    j=pj
    while True:
      if tab[i][j]==peca:
        ok=True
        mostratab(i,j,"x")
        break
      i+=direcaoi[k]
      j+=direcaoj[k]
      if i==limitei[k] or j==limitej[k]:
        break
    if ok:
      break
  return ok

def vRainha(pi,pj,peca):
  if vTorre(pi,pj,peca) or vBispo(pi,pj,peca):
    return True
  else:
    return False

posicaoi=int(input("Entre com a linha da peca a ser analisada:"))
posicaoj=int(input("Entre com a coluna da peca a ser analisada:"))
if vBispo(posicaoi,posicaoj,'b'):
  print("Ameaçada pelo bispo")
if vTorre(posicaoi,posicaoj,'t'):
  print("Ameaçada pelo torre")
if vRainha(posicaoi,posicaoj,'r'):
  print("Ameaçada pela Rainha")

````
