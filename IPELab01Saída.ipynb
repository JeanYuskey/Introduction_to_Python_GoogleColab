É possivel formatar o próprio número:
especificando a quantidade de dígitos reservados.
_d_ no caso para números inteiros, _7d_ indica 7 dígitos inteiros.

```python
print("valor de A={0} e B={1} certo?" .format(a,b))
print("valor de A={0:7d} e B={1} certo?" .format(a,b))
```

Também o seu alinhamento

```python
print("valor de A={0} e B={1} certo?".format(a,b))
print("valor de A={0:7d} e B={1} certo?".format(a,b))
print("valor de A={0:<7d} e B={1} certo?".format(a,b))  #esquerda
print("valor de A={0:^7d} e B={1} certo?".format(a,b))  #centralizado
```

#Comentários

* Comentários são trechos do programa voltados para a leitura por humanos, e ignorados pelo interpretador

* Começam com o símbolo **#**

* Tudo na linha após **#** e ignorado pelo interpretador

* Use comentários para documentar seu código e fazer com que ele seja fácil de entender por outras pessoas



```python
# Corrija a indentação
lista = ['laranja', 'banana', 'uva']
lista.sort()
for x in lista:
  print (x.capitalize())
```
