1. **Criação do arquivo**


```python
opcao = int(input('1.Criar\t2.Gravar\t3.Abrir: ')) #\t é tab(indentação)
nome =  input('Nome do arquivo: ')
nome += '.txt' #nome+='.txt' transforma a variável nome em um arquivo txt

if opcao == 1:
  arquivo = open(nome,'w',encoding='utf-8') #utilizar o modo 'w' primeiro para a criação do arquivo, depois pode utilizar o 'a' e #encoding= ' utf-8' serve como seguro, garantia para o arquivo poder ser lido em texto
  import random
  for i in range (10):
    arquivo.write(str(random.randint(0,1000))+'\n') # \n é quebra de linha(contatenação)
  arquivo.close()

elif opcao == 2:
  arquivo = open(nome,'a',encoding='utf-8')
  import random
  for i in range (10):
    arquivo.write(str(random.randint(0,1000))+'\n') # \n é quebra de linha(contatenação)
  arquivo.close()

elif opcao == 3:
  arquivo = open(nome,'r',encoding='utf-8') #readlines tranforma o arquivo em uma lista
  print(arquivo) #arquivo é a variável e o arquivo2 é uma lista em relação a variável arquivo
  arquivo2 = arquivo.readlines()
  print(arquivo2)
  nova=[]
  for i in arquivo2:
    print(i.rstrip()) #.rstrip(), que remove caracteres de marcação (espaço, quebra de linha) e os caracteres indicados no parêntese [TEXTO]
    nova.append(int(i.rstrip()))
  print(nova)
  arquivo.close()
else:
  print('Tá errado, mano!')
```

**Modos de abrir um arquivo:**

- **“w” (write):** cria um arquivo ou abre um arquivo já existente para ser reescrito; quaisquer dados presentes no arquivo antes da execução desta instrução são apagados;

- **“r” (read):** abre uma arquivo para leitura; embora todos os dados estejam presentes para serem usados no programa, não permite gravação no arquivo; é modo que é utilizado, se nenhum for especificado;

- **“a” (append):** permite inserir instruções ao final de um arquivo; caso os dados precisem ser inseridos em outro ponto, devemos acessar o arquivo com o modo de leitura, organizar os dados e então inseri-los (usando os modo “w” ou “a”).

Após encerrar-se o acesso a uma arquivo, precisamos fechá-lo usando **arquivo.close();**

Para garantir a correta conversão dos caracteres do arquivo para o Python e vice-versa, podemos utilizar o parâmetro encoding=“utf-8“ no método open:

**variável=open("arquivo","modo",encoding="utf-8")**

Para gravarmos um dado em um arquivo, utilizamos o método .**write()**, colocando no parênteses o que será escrito no arquivo;


**Exibição do arquivo aberto**

Caso nada seja especificado, os dados serão todos concatenados; para que haja uma quebra de linha **(\n)** ou algum separador, devemos especificar isto, acrescentado o separador junto com o conteúdo que será inserido.

Para exibirmos um arquivo, é necessário utilizar o método .**readlines()**, que irá transformar o arquivo em uma lista dentro do programa;

Para que o separador do elementos do arquivo não sejam exibidos, podemos utilizar o método **.rstrip()**, que remove caracteres de marcação (espaço, quebra de linha) e os caracteres indicados no parêntese 

```python
nome=input("Nome do Arquivo: ")
arq= open(nome, "w")
print("Arquivo criado")
arq.close()

arq= open(nome, "a")
texto=input("Digite um texto a ser inserido: ")
arq.write(texto.lower()+"\n")
arq.close()

arq= open(nome, "r",encoding="utf-8")
arq=arq.readlines()
print(arq)
for linha in arq:
  linha = linha.rstrip()
  print (linha)
```
